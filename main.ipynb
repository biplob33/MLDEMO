```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
from sklearn.linear_model import LinearRegression,ElasticNet
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_dataset():
    x = np.array([[random.randint(0,2)], [random.randint(2,5)], [random.randint(6,10)], [random.randint(10,12)]])
    y = np.array([[random.randint(0,2)], [random.randint(2,5)], [random.randint(6,10)], [random.randint(10,12)]])
    return x,y
```

```python
def update():
    plt.close
    x,y = generate_dataset()
    lin = LinearRegression().fit(x, y)
    l1 = lin.predict([x[0]]).reshape(1,-1)
    l2 = lin.predict([x[-1]]).reshape(1,-1)
    els = ElasticNet(random_state=0).fit(x, y)
    e1 = els.predict([x[0]])
    e2 = els.predict([x[-1]])
    plt.scatter(x,y)
    plt.plot([x[0],x[-1]],[l1[0],l2[0]])
    plt.plot([x[0],x[-1]],[e1[0],e2[0]])
```

```python
button = widgets.Button(description="Click Me!")
display(button)
def on_button_clicked(b):
    update()
button.on_click(on_button_clicked)
```
